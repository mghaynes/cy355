### The Shell

````python
from IPython.display import display, Markdown
import ipywidgets as widgets

shell_q1 = widgets.HTML(value='<b>Shell scripts are collections of shell _____________</b>')
shell_answer1 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

shell_q2 = widgets.HTML(value='<b>The shell is a text-based interface to the _____________</b>')
shell_answer2 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

show_shell_answers = widgets.Button(description='Show Answers')
hide_shell_answers = widgets.Button(description='Hide Answers')
shell_answers_output = widgets.Output()

def show_shell(b):
    with shell_answers_output:
        display(Markdown('''
        - Shell scripts are collections of shell **commands**.
        - The shell is a text-based interface to the **operating system**.
        '''))

def hide_shell(b):
    shell_answers_output.clear_output()

show_shell_answers.on_click(show_shell)
hide_shell_answers.on_click(hide_shell)

display(shell_q1)
display(shell_answer1)
display(shell_q2)
display(shell_answer2)
display(show_shell_answers)
display(hide_shell_answers)
display(shell_answers_output)
````

### Managing the System

````python
managing_q1 = widgets.HTML(value='<b>Daemons are _____________ that run in the background to manage different subsystems.</b>')
managing_answer1 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

managing_q2 = widgets.HTML(value='<b>A user uses the systemctl command to interface with _____________</b>')
managing_answer2 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

show_managing_answers = widgets.Button(description='Show Answers')
hide_managing_answers = widgets.Button(description='Hide Answers')
managing_answers_output = widgets.Output()
       
def show_managing(b):
    with managing_answers_output:
        display(Markdown('''
        - Daemons are **processes** that run in the background to manage different subsystems.
        - A user uses the systemctl command to interface with **system services**.
        '''))

def hide_managing(b):
    managing_answers_output.clear_output()

show_managing_answers.on_click(show_managing)
hide_managing_answers.on_click(hide_managing)

display(managing_q1)
display(managing_answer1)
display(managing_q2)
display(managing_answer2)
display(show_managing_answers)
display(hide_managing_answers)
display(managing_answers_output)
````

### The Filesystem

````python
filesystem_q1 = widgets.HTML(value='<b>The Linux file system has a _____________ structure.</b>')
filesystem_answer1 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

filesystem_q2 = widgets.HTML(value='<b>A file that contains a set of machine instructions is a _____________.</b>')
filesystem_answer2 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

filesystem_q3 = widgets.HTML(value='<b>Machine instructions that are in a state of execution are a _____________</b>')
filesystem_answer3 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

filesystem_q4 = widgets.HTML(value='<b>The three permissions groups in the Linux filesystem are: _____________, _____________, and _____________</b>')
filesystem_answer4 = widgets.Textarea(value='', placeholder='Enter your answer', layout=widgets.Layout(width='100%'))

show_filesystem_answers = widgets.Button(description='Show Answers')
hide_filesystem_answers = widgets.Button(description='Hide Answers')
filesystem_answers_output = widgets.Output()

def show_filesystem(b):
    with filesystem_answers_output:
        display(Markdown('''
        - The Linux file system has a **hierarchical** structure.
        - A file that contains a set of machine instructions is a **binary executable**.
          (hint: we would find these in /bin or /usr/bin)
        - Machine instructions that are in a state of execution are a **process**.
          (hint: we would find these in /proc)
        - The three permissions groups in the Linux filesystem are: **User**, **Group**, and **Others**.
        '''))

def hide_filesystem(b):
    filesystem_answers_output.clear_output()

show_filesystem_answers.on_click(show_filesystem)
hide_filesystem_answers.on_click(hide_filesystem)

display(filesystem_q1)
display(filesystem_answer1)
display(filesystem_q2)
display(filesystem_answer2)
display(filesystem_q3)
display(filesystem_answer3)
display(filesystem_q4)
display(filesystem_answer4)
display(show_filesystem_answers)
display(hide_filesystem_answers)
display(filesystem_answers_output)
````
